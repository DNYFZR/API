```python
import pandas as pd
pd.set_option('display.max_rows', 500)
```

```python
# Query API
url = r'''
http://localhost/ATP_tour_data?table=matches&
date_col=tourney_date&
date_from=2005&
tournament_col=tourney_name&
tournament=roland_garros&
player_col=winner_name&
player_col=loser_name&
player=rafael_nadal&
select_col=winner_name&
select_col=loser_name&
select_col=round&&
select_col=score&
select_col=match_num&
select_col=tourney_date'''.replace('\n', '')

df = pd.read_json(url)
```

```python
# Extract year from dates 
df['tourney_date'] = pd.to_datetime(df['tourney_date']).dt.year

# Sort by year and round
df = df.sort_values(by=['tourney_date', 'match_num']).reset_index(drop=True)

df 
```
